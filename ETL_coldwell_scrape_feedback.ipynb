```python
# Browser going to the page to be scraped
browser_bot.visit('https://www.coldwellbanker.com/for-sale-homes/Beaverton-OR-989c/beds_2/baths_2')
```

```python
coldwell_html = browser_bot.html
```

```python
# Scraper bot
coldwell_scraper = bs(coldwell_html, 'html.parser')
coldwell_scraper
```

```python
# Identify the info we need (class directory up) #(TO DELETE LATER: l-two-col-main, sr-card__street, media__content)
all_listing_info = coldwell_scraper.findAll('div', {'class': 'l-two-col-main'})
all_listing_info
```

```python
listings = []

for listing in all_listing_info:
    
    # Fetch all values desired
#    values = listing.find_all("div", {"class": "media__content"})
    values = listing.find_all("div", {"class": "sr-card__prop-details"})
    val_text = listing.get_text()
    print(val_text)
#    print(len(values))
#    print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-") 

    for value in values:
         
        # Fetch Street Address
#       PREVIOUS VERSION: street_address = listing.get_text("div", {"itemprop":"streetAddress"})
        street_address = listing.get_text("div", {"itemprop":"address"})
#       !!! IF LINE IMMEDIATELY ABOVE GOES FROM ".get_text" to ".find_all", THE PRINT STATEMENT IMMEDIATELY 
#       !!! BELOW SHOWS ALL TEXT FOR EACH LISTING, GROUPED TOGETHER AS A PARAGRAPH
        print((street_address) + str("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*--*-*-"))
        
        # Fetch Cities
        all_cities = listing.get_text("span", {"itemprop":"addressLocality"})
    
        # Fetch state
        all_states = listing.get_text("span", {"itemprop":"addressRegion"})
    
        # Fetch all zips
        all_zip = listing.get_text('span', {'itemprop': 'postalCode'})
    
        # Fetch all prices
        all_price = listing.get_text('span', {'class': 'font-list-price font-20'})

        # Fetch all HOA
        all_hoa = listing.get_text('div', {'class': 'l-grid__col-1-5 f-right t-right font-666'})
    
        # Fetch all bedroom counts
        all_beds = listing.get_text('li', {'class': 'test-beds'})
    
        # Fetch all bath counts
        all_baths = listing.get_text('li', {'class': 'test-baths'})
    
        # Fetch all sqft
        all_sqft = listing.get_text('div', {'class': 'sr-card__prop-details'})
    
        # Fetch all price per sqft
        all_ppsqft = listing.get_text('div', {'class': 'sr-card__prop-details'})
    
        # Fetch all days listed
        all_days_listed = listing.get_text('div', {'class': 'sr-card__prop-details mt-0'})
    
        # Fetch all home type
        all_home_type = listing.get_text('div', {'class': 'sr-card__prop-details mt-0'})
        
        
        listing_object = {
                        'Street Address': all_street_address,
                        'City': all_cities,
                        'State': all_states,
                        'Zip Code': all_zip,
                        'Price': all_price,
                        'HOA Fees': all_hoa,
                        'Bedroom Counts': all_beds,
                        'Bathroom Counts': all_baths,
                        'SqFt': all_sqft,
                        'Price Per SqFt': all_ppsqft,
                        'Days Listed': all_days_listed,
                        'Home Type': all_home_type
                        }

    # Append all scraped values to list
    listings.append(listing_object)   
```
